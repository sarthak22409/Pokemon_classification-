```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(centers=2,n_samples=500,n_features = 2,random_state=4)
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap="coolwarm")
```

```python
def sigmoid(z):
    return 1/(np.exp(-z)+1)
```

```python
def predict(X,weights):
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions
def loss(X,Y,weights):
    Y_= predict(X,weights)
    cost = np.mean(-Y*np.log2(Y_) - (1-Y)*np.log2(1-Y_))
    return cost

def update(X,Y,weights,learning_rate):
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_-Y)
    m = X.shape[0]
    weights = weights - learning_rate*dw/m
    return weights

def train(X,Y,learning_rate=0.5,max_epochs = 100):
    ones = np.ones((X.shape[0],1))
    losses = []
    X=np.hstack((ones,X))
    weights = np.zeros(X.shape[1])
    for epoch in range( max_epochs):
        
        l = loss(X,Y,weights)
        weights = update(X,Y,weights,learning_rate)
        losses.append(l)
        print("Epoch %d loss %.4f"%(epoch,l))
            
    return weights,losses
```

```python
weights,loss = train(X,Y,learning_rate=0.8,max_epochs=1000)
```

```python
def get_predictions(X,weights,labels):
    if(X.shape[1]!=weights.shape[0]):
        ones = np.ones((X.shape[0],1))
        X_test=np.hstack((ones,X))
        
    probs = predict(X_test,weights)
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs>=0.5]=1
        return labels
```
